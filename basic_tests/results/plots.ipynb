```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
files = ["test-zlib-1.csv", "test-zlib-2.csv", "test-zlib-3.csv", "test-zlib-4.csv"]
dfs = []
for file in files:
    dfs.append(pd.read_csv(file)) 
    
dfs[0]
```

```python
# Split mean ce into quartiles
dfs[0]['Mean CE Quartile'] = pd.qcut(dfs[0]['Mean CE'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
dfs[1]['Mean CE Quartile'] = pd.qcut(dfs[1]['Mean CE'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
dfs[2]['Mean CE Quartile'] = pd.qcut(dfs[2]['Mean CE'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
dfs[3]['Mean CE Quartile'] = pd.qcut(dfs[3]['Mean CE'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
```

```python

sns.scatterplot(data=dfs[0], x='Bits Truncated', y='Mean CE', hue='Mean CE Quartile')
```
