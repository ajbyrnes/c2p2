```python

def print_branches(branches):
    for branch in branches:
        print(f'{branch}: {tree[branch].array()}')
    
def print_unique_values(branches):
    for branch in branches:
        branch_arr = tree[branch].array()
        if branch_arr.ndim > 1:
            branch_arr = ak.flatten(branch_arr)

        print(f'{branch}: {np.unique(branch_arr)}')

def get_colors(branches, start='rgb(5, 200, 200)', end='rgb(200, 10, 10)'):
    return n_colors(start, end, len(branches), colortype='rgb')

def ridgeline(branches, colors=None, name='all'):
    fig = go.Figure()

    if colors is None:
        colors = get_colors(branches)

    for branch, color in zip(branches, colors):
        branch_arr = tree[branch].array()
        if branch_arr.ndim > 1:
            branch_arr = ak.flatten(branch_arr)

        fig.add_trace(
            go.Violin(
                x = branch_arr,
                name = branch,
                orientation = 'h',
                side = 'positive',
                points = False,
                line_color = color,
                width = 3,
            )
        )

    fig.update_layout (
        xaxis_showgrid = False,
        xaxis_zeroline = False,
    )

    fig.write_html(f'ridgeline_{name}.html')
```

```python
ridgeline(scale_branches, name='scaleFactor')
ridgeline(lep_branches, name='lep')
ridgeline(largejet_branches, name='largeRjet')
ridgeline(jet_branches, name='jet')
ridgeline(met_branches, name='met')
```
