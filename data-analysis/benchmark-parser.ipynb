```python
import re
import pandas as pd
```

```python
# Template for bit truncation logs
# Compressor: Trunk
# Iterations: 5
# Data name: root
# Source file: mc_361106.Zee.1largeRjet1lep.root
# Tree name: mini
# Branch name: jet_E
# Precision: 3
# Trunk compression level: 9
# Average compression time: 103.80 ms (user: 104.00 ms, system: 0.00 ms)
# Average decompression time: 1.20 ms (user: 2.00 ms, system: 0.00 ms)
# Average compression memory: 0 KB
# Original data size: 535336 bytes
# Compressed data size: 285167 bytes
# Compression ratio: 1.88
# Average relative error: 0.000176
# ==========
```

```python
# Template for SZ3 logs
# Compressor: SZ
# Iterations: 5
# Data name: root
# Source file: mc_361106.Zee.1largeRjet1lep.root
# Tree name: mini
# Branch name: jet_E
# Precision: 3
# SZ error bound mode: 1
# SZ algorithm: 0
# SZ interpolation algorithm: 1
# Average compression time: 31.60 ms (user: 32.00 ms, system: 0.00 ms)
# Average decompression time: 23.20 ms (user: 24.00 ms, system: 0.00 ms)
# Average compression memory: 0 KB
# Original data size: 535336 bytes
# Compressed data size: 136665 bytes
# Compression ratio: 3.92
# Average relative error: 0.014123
# ==========
```

```python
header = [
    "Compressor",
    "Iterations",
    "Data name",
    "Source file",
    "Tree name",
    "Branch name",
    "Precision",
    "Average compression time (ms)",
    "Average decompression time (ms)",
    "Original data size (bytes)",
    "Compressed data size (bytes)",
    "Compression ratio",
    "Average relative error",
    "Trunk compression level",
    "SZ error bound mode",
    "SZ algorithm",
    "SZ interpolation algorithm"
]

def parse_log(logfile): 
    # Define common regex patterns
    compressor_re = r"Compressor:\s*(\w+)"
    iterations_re = r"Iterations:\s*(\d+)"
    data_name_re = r"Data name:\s*(\w+)"
    source_file_re = r"Source file:\s*([\w.]+)"
    tree_name_re = r"Tree name:\s*([\w.]+)"
    branch_name_re = r"Branch name:\s*([\w.]+)"
    precision_re = r"Precision:\s*(\d+)"
    
    average_compression_time_re = r"Average compression time:\s*([\d.]+)\s*ms"
    average_decompression_time_re = r"Average decompression time:\s*([\d.]+)\s*ms"
    original_data_size_re = r"Original data size:\s*([\d.]+)\s*bytes"
    compressed_data_size_re = r"Compressed data size:\s*([\d.]+)\s*bytes"
    compression_ratio_re = r"Compression ratio:\s*([\d.]+)"
    average_relative_error_re = r"Average relative error:\s*([\d.]+)"
    
    # Define truncation regex patterns
    trunk_compression_level_re = r"Trunk compression level:\s*(\d+)"
    
    # Define SZ3 regex patterns
    sz_error_bound_mode_re = r"SZ error bound mode:\s*(\d+)"
    sz_algorithm_re = r"SZ algorithm:\s*(\d+)"
    sz_interpolation_algorithm_re = r"SZ interpolation algorithm:\s*(\d+)"
    
    # Read the log file
    with open(logfile, 'r') as f:
        log_content = f.read()
        
    # Parse log content into entries
    delimiter = "=========="
    log_entries = log_content.split(delimiter)
    
    # Parse each entry
    rows = []
    for entry in log_entries:
        if not entry.strip():
            continue
        
        # Extract values using regex
        compressor = re.search(compressor_re, entry)
        iterations = re.search(iterations_re, entry)
        data_name = re.search(data_name_re, entry)
        source_file = re.search(source_file_re, entry)
        tree_name = re.search(tree_name_re, entry)
        branch_name = re.search(branch_name_re, entry)
        precision = re.search(precision_re, entry)
        
        average_compression_time = re.search(average_compression_time_re, entry)
        average_decompression_time = re.search(average_decompression_time_re, entry)
        original_data_size = re.search(original_data_size_re, entry)
        compressed_data_size = re.search(compressed_data_size_re, entry)
        compression_ratio = re.search(compression_ratio_re, entry)
        average_relative_error = re.search(average_relative_error_re, entry)
        
        trunk_compression_level = re.search(trunk_compression_level_re, entry)
        
        sz_error_bound_mode = re.search(sz_error_bound_mode_re, entry)
        sz_algorithm = re.search(sz_algorithm_re, entry)
        sz_interpolation_algorithm = re.search(sz_interpolation_algorithm_re, entry)

        # Create row for current log entry
        row = [
            compressor.group(1) if compressor else None,
            int(iterations.group(1)) if iterations else None,
            data_name.group(1) if data_name else None,
            source_file.group(1) if source_file else None,
            tree_name.group(1) if tree_name else None,
            branch_name.group(1) if branch_name else None,
            int(precision.group(1)) if precision else None,
            float(average_compression_time.group(1)) if average_compression_time else None,
            float(average_decompression_time.group(1)) if average_decompression_time else None,  # Average compression memory (not provided in the log)
            int(original_data_size.group(1)) if original_data_size else None,
            int(compressed_data_size.group(1)) if compressed_data_size else None,
            float(compression_ratio.group(1)) if compression_ratio else None,
            float(average_relative_error.group(1)) if average_relative_error else None,
            int(trunk_compression_level.group(1)) if trunk_compression_level else -1,
            int(sz_error_bound_mode.group(1)) if sz_error_bound_mode else -1,
            int(sz_algorithm.group(1)) if sz_algorithm else -1,
            int(sz_interpolation_algorithm.group(1)) if sz_interpolation_algorithm else -1,
        ]
        
        rows.append(row)
        
    return rows
```

```python
name = "2025-05-02-results"
logfile = f"{name}.log"
csvfile = f"{name}.csv"

rows = parse_log(logfile)

pd.DataFrame(rows, columns=header).to_csv(csvfile, index=False)
print(f"Parsed log file '{logfile}' and saved to '{csvfile}'")
```
